```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Optional: give the displayed data cleaner formatting

```python
#Extract the columns
complete_data = school_data_complete.loc[:, ['school_name','type','student_name',
                                         'reading_score','math_score','budget']]
complete_data
```

```python
math_df=complete_data.groupby(['school_name','type'])['math_score'].apply(lambda x: x[x >=70].count())
math_df
```

```python
math_final=pd.DataFrame(math_df)
math_final=math_final.rename(
    columns={"math_score":"math_pass"})
math_final
```

```python
reading_df=complete_data.groupby(['school_name','type'])['reading_score'].apply(lambda x: x[x >=70].count())
reading_df
```

```python
reading_final=pd.DataFrame(reading_df)
reading_final=reading_final.rename(
    columns={"reading_score":"reading_pass"})
reading_final
```

```python
Overall_df=complete_data.loc[(complete_data['reading_score']>=70) & (complete_data['math_score']>=70)]
Overall_df
```

```python
Overall_pass=Overall_df.groupby(['school_name','type'])['reading_score'].apply(lambda x: x[x >=70].count())
Overall_pass
```

```python
Overall_final=pd.DataFrame(Overall_pass)
Overall_final=Overall_final.rename(
    columns={"reading_score":"overall_pass"})
Overall_final
```

```python
Total_Score_df=pd.merge(reading_final,math_final,on='school_name')
Total_Score_df
```

```python
Total_Score_final=Total_Score_df.merge(Overall_final,on='school_name')
Total_Score_final
```

```python
#Create a dataframe and group by school name
complete_data_df= complete_data.groupby(['school_name']).aggregate({'student_name':'count',
                                                                    'reading_score':'sum',
                                                                    'math_score':'sum',
                                                                    'budget':'mean'})


complete_data_df
```

```python
complete_data_df=complete_data_df.merge(Total_Score_final, on='school_name')
complete_data_df
```

```python
val1=len(complete_data_df)
val2=sum(complete_data_df['student_name'])
val3=sum(complete_data_df['budget'])
val4=float(sum(complete_data_df['math_score'])/val2)
val5=float(sum(complete_data_df['reading_score'])/val2)
val6=float((sum(complete_data_df['math_pass'])/val2))
val7=float((sum(complete_data_df['reading_pass'])/val2))
val8=float((sum(complete_data_df['overall_pass'])/val2))
Summary_final=pd.DataFrame({'index':[0],
                            'Total Schools':val1,
                            'Total Students':val2,
                            'Total Budget':val3,
                            'Average Math Score':val4,
                            'Average Reading Score':val5,
                            '% Passing Math':val6,
                            '% Passing Reading':val7,
                            '% Overall Passing':val8
                           })
```

```python
del Summary_final['index']
```

```python
Summary_final['Total Students']=Summary_final['Total Students'].apply('{:,}'.format)
Summary_final['Total Budget']=Summary_final['Total Budget'].apply('${0:,.2f}'.format)
Summary_final['% Passing Math']=Summary_final['% Passing Math'].apply('{:.2%}'.format)
Summary_final['% Passing Reading']=Summary_final['% Passing Reading'].apply('{:.2%}'.format)
Summary_final['% Overall Passing']=Summary_final['% Overall Passing'].apply('{:.2%}'.format)
Summary_final
```
